```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_json(r"https://www.mohfw.gov.in/data/datanew.json")

# r before the string
# represent that the string is a raw string
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.sort_values("active")
```

```python
x = data.drop(axis=0, labels=36)
```

```python
y = x.sort_values("active").tail(7)
```

```python
y
```

## Pie Chart of Active Cases

Let's plot a pie chart to visualize the 7 states with active cases of Covid

```python

plt.pie(y.active, labels=y.state_name)
plt.title(" Top 7 States with Active Cases")
plt.show()
```

```python
c = x.sort_values("cured").tail(7)
```

```python
c
```

## Pie Chart of Most Cured Reports

Let's plot a chart that shows the top 7 states with highest Cured reports

```python
plt.pie(c.cured, labels=c.state_name)
plt.title("Pie Chart of Cured cases", color="Brown")
plt.show()
```

## Visualise maximum death cases 


```python
d = x.sort_values("death").tail(7)

# colors = ["orange","red","indigo","cyan"]
plt.pie(d.death, labels=d.state_name)
plt.title("Pie Chart of Max Death Cases")
plt.show()
```
